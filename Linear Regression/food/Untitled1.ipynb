```python
import pandas as pd
```

```python
df=pd.read_csv("Food_Delivery_Times.csv")
df.head()
```

```python
df=df.drop(columns='Order_ID')
```

```python
df = df.dropna()
df.shape
```

```python
df_encoded = pd.get_dummies(df, columns=["Weather","Traffic_Level","Time_of_Day","Vehicle_Type"], drop_first=False)
df_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
X = df_encoded.drop(columns=["Delivery_Time_min"])  
y = df_encoded["Delivery_Time_min"] 
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
df_encoded.to_csv("Saket_cleaned_data_final.csv", index=False)
```
